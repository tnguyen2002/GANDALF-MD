```python
import src
from PIL import Image, ImageDraw, ImageFont
from src.datasets import cifar10

from torchvision import transforms
from PIL import ImageFilter, Image
import random
```

```python
class GaussianBlur(object):
    """Gaussian blur augmentation in SimCLR https://arxiv.org/abs/2002.05709"""

    def __init__(self, sigma=(0.1, 2.0)):
        self.sigma = sigma

    def __call__(self, x):
        sigma = random.uniform(self.sigma[0], self.sigma[1])
        x = x.filter(ImageFilter.GaussianBlur(radius=sigma))
        return x

train_transforms = transforms.Compose([
            transforms.RandomResizedCrop(32, scale=(0.2, 1.)),
            transforms.RandomApply([
                transforms.ColorJitter(0.4, 0.4, 0.4, 0.1)  # not strengthened
            ], p=0.8),
            transforms.RandomGrayscale(p=0.2),
            transforms.RandomApply([GaussianBlur([.1, 2.])], p=0.5),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            #transforms.Normalize(mean=[0.491, 0.482, 0.446],
            #                    std=[0.247, 0.243, 0.261]),
        ])

test_transforms = transforms.Compose([
            transforms.Resize(32),
            transforms.CenterCrop(32),
            transforms.ToTensor(),
        ])
```

```python
dataset = cifar10.CIFAR10Letters(image_transforms = test_transforms, alternate_label=True)

import matplotlib.pyplot as plt
for idx in range(10):
    plt.figure()
    plt.title(dataset[idx][-1])
    plt.imshow(dataset[idx][1].permute(1,2,0))
```
