```python
from torchvision import transforms
from PIL import ImageFilter, Image
import random

class GaussianBlur(object):
    """Gaussian blur augmentation in SimCLR https://arxiv.org/abs/2002.05709"""

    def __init__(self, sigma=(0.1, 2.0)):
        self.sigma = sigma

    def __call__(self, x):
        sigma = random.uniform(self.sigma[0], self.sigma[1])
        x = x.filter(ImageFilter.GaussianBlur(radius=sigma))
        return x

train_transforms = transforms.Compose([
            transforms.RandomResizedCrop((32,32), scale=(0.2, 1.)),
            transforms.RandomApply([
                transforms.ColorJitter(0.4, 0.4, 0.4, 0.1)  # not strengthened
            ], p=0.8),
            transforms.RandomGrayscale(p=0.2),
            transforms.RandomApply([GaussianBlur([.1, 2.])], p=0.5),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            #transforms.Normalize(mean=[0.491, 0.482, 0.446],
            #                    std=[0.247, 0.243, 0.261]),
        ])

test_transforms = transforms.Compose([
            transforms.Resize((32,32)),
            #transforms.CenterCrop(32),
            transforms.ToTensor(),
        ])



dataset = ham.HAM(image_transforms = train_transforms)
#dataset = pannuke.PanNuke(image_transforms = train_transforms)
#dataset = idrid.IDRiD(image_transforms = train_transforms)
```

```python
import os
from src.utils import utils
import dotmap
from dotmap import DotMap
from src.systems.image_systems import PretrainViewMakerSystem
from src.systems.image_systems import PretrainExpertSystem
import torch
import matplotlib.pyplot as plt
import math
from mpl_toolkits.axes_grid1 import ImageGrid
from matplotlib.colors import BoundaryNorm

#pretrain_viewmaker_ham_budget_0_00625
#pretrain_viewmaker_pannuke_budget_0_05
#pretrain_viewmaker_idrid_rev2_budget_0_1
#pretrain_viewmaker_idrid_budget_0_025

base_dir = '/mnt/fs0/atamkin/feature_suppression_xiluo/experiments/pretrain_viewmaker_ham_budget_0_00625/'
config_path = os.path.join(base_dir, 'config.json')
config_json = utils.load_json(config_path)
config = DotMap(config_json)
SystemClass = globals()[config.system]
system = SystemClass(config)
checkpoint_file = os.path.join(base_dir, 'checkpoints', 'epoch=199.ckpt')
system.load_state_dict(torch.load(checkpoint_file, map_location='cpu')['state_dict'])

grid_size = 3

for img_idx in range(20):

    # Grab a random example
    example = system.train_dataset[img_idx][1]
    print('label: ' + str(system.train_dataset[img_idx][4]))
    example = example.unsqueeze(0)

    # Assemble the grid.
    
    #VIEWMAKER
    grid_imgs = []
    grid_imgs_diff = []
    for _ in range(grid_size ** 2):
        if _ == math.floor((grid_size ** 2) / 2):
            grid_imgs.append(example)
            grid_imgs_diff.append(example)
        else:
            view = system.viewmaker(example)
            grid_imgs.append(view)
            diff = example - view
            grid_imgs_diff.append(diff)
                                 
    fig2 = plt.figure(figsize=(grid_size, grid_size))
    grid2 = ImageGrid(fig2, 111,  # similar to subplot(111)
                     # creates grid_size x grid_size grid of axes
                     nrows_ncols=(grid_size, grid_size),
                     axes_pad=0,  # pad between axes in inch.
                     share_all=True,
                     )

    
    for ax, im in zip(grid2, grid_imgs):
        #ax.set_xticks([])
        #ax.set_yticks([])
        ax.tick_params(bottom = False, left = False)
        # Remove batch dim and put channel dim last.
        ax.imshow(im[0].detach().permute(1, 2, 0))
    
    grid2[0].get_yaxis().set_ticks([])
    grid2[0].get_xaxis().set_ticks([])
    plt.axis('off')
    
    #plt.savefig('./gridimgnew/letters_viewmaker_' + str(img_idx) + '.jpg',bbox_inches='tight', pad_inches=0.0)
    plt.show()
    plt.clf()
    
    #VIEWMAKER_DIFF

    low0 = 0
    high0 = 0
    low1 = 0
    high1 = 0
    low2 = 0
    high2 = 0
    for i, im in enumerate(grid_imgs_diff):
        if i != math.floor((grid_size ** 2) / 2):
            low0 = min([low0, torch.min(im[0, 0]).detach()])
            high0 = max([high0, torch.max(im[0, 0]).detach()])
            low1 = min([low1, torch.min(im[0, 1]).detach()])
            high1 = max([high1, torch.max(im[0, 1]).detach()])
            low2 = min([low2, torch.min(im[0, 2]).detach()])
            high2 = max([high2, torch.max(im[0, 2]).detach()])
    
    low = min(min(low0, low1), low2)
    high = max(max(high0, high1), high2)
            
    
    for i, im in enumerate(grid_imgs_diff):
        if i != math.floor((grid_size ** 2) / 2):
            #im[0, 0] = (im[0, 0] - low0) / (high0 - low0)
            #im[0, 1] = (im[0, 1] - low1) / (high1 - low1)
            #im[0, 2] = (im[0, 2] - low2) / (high2 - low2)
            #im[0, 0][im[0, 0] <= 0.55] = 0
            #im[0, 1][im[0, 1] <= 0.55] = 0
            #im[0, 2][im[0, 2] <= 0.55] = 0
            
            im[0, 0] = (im[0, 0] / 2) + 0.5
            im[0, 1] = (im[0, 1] / 2) + 0.5
            im[0, 2] = (im[0, 2] / 2) + 0.5
    fig = plt.figure(figsize=(grid_size, grid_size))
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                     # creates grid_size x grid_size grid of axes
                     nrows_ncols=(grid_size, grid_size),
                     axes_pad=0,  # pad between axes in inch.
                     share_all=True,
                     )
    
    for ax, im in zip(grid, enumerate(grid_imgs_diff)):
        i, im = im
        #ax.set_xticks([])
        #ax.set_yticks([])
        ax.tick_params(bottom = False, left = False)
        # Remove batch dim and put channel dim last.
        if i == math.floor((grid_size ** 2) / 2):
            ax.imshow(im[0].detach().permute(1, 2, 0))
        else:
            ax.imshow(im[0].detach().permute(1, 2, 0))
            #ax.imshow(im[0].detach().permute(1, 2, 0), norm=norm, cmap=cmap)
    
    grid[0].get_yaxis().set_ticks([])
    grid[0].get_xaxis().set_ticks([])
    plt.axis('off')
    
    
    #plt.savefig('./gridimgnew/letters_viewmakerdiff_' + str(img_idx) + '.jpg',bbox_inches='tight', pad_inches=0.0)
    plt.show()
    plt.clf()
    
    grid_imgs_expert = []
    for _ in range(grid_size ** 2):
        if _ == math.floor((grid_size ** 2) / 2):
            grid_imgs_expert.append(example)
        else:
            grid_imgs_expert.append([dataset[img_idx][1]])
    
   
    fig1 = plt.figure(figsize=(grid_size, grid_size))
    grid1 = ImageGrid(fig1, 111,  # similar to subplot(111)
                     # creates grid_size x grid_size grid of axes
                     nrows_ncols=(grid_size, grid_size),
                     axes_pad=0,  # pad between axes in inch.
                     share_all=True,
                     )

    
    for ax, im in zip(grid1, grid_imgs_expert):
        #ax.set_xticks([])
        #ax.set_yticks([])
        ax.tick_params(bottom = False, left = False)
        # Remove batch dim and put channel dim last.
        ax.imshow(im[0].detach().permute(1, 2, 0))
    
    grid1[0].get_yaxis().set_ticks([])
    grid1[0].get_xaxis().set_ticks([])
    plt.axis('off')
    
    #plt.savefig('./gridimgnew/letters_expert_' + str(img_idx) + '.jpg',bbox_inches='tight', pad_inches=0.0)
    plt.show()
    plt.clf()
```

```python
import os
from src.utils import utils
import dotmap
from dotmap import DotMap
from src.systems.audio_systems import PretrainViewmakerAudioMNISTSystem
from src.systems.audio_systems import PretrainExpertAudioMNISTSystem
import torch
import matplotlib.pyplot as plt
import math
from mpl_toolkits.axes_grid1 import ImageGrid
from matplotlib.colors import BoundaryNorm

base_dir = '/mnt/fs1/atamkin/feature_suppression_xiluo/experiments/pretrain_viewmaker_audiomnist_simclr_L1_forced_b01_100hour_64x64_resnet18_full_noise005_budget0125_seed2/'
config_path = os.path.join(base_dir, 'config.json')
config_json = utils.load_json(config_path)
config = DotMap(config_json)
SystemClass = globals()[config.system]
system = SystemClass(config)
checkpoint_file = os.path.join(base_dir, 'checkpoints', 'epoch=199.ckpt')
system.load_state_dict(torch.load(checkpoint_file, map_location='cpu')['state_dict'])

grid_size = 3

for img_idx in range(10):

    # Grab a random example
    example = system.train_dataset[img_idx][1]
    example = example.unsqueeze(0)

    # Assemble the grid.
    grid_imgs = []
    for _ in range(grid_size ** 2):
        if _ == math.floor((grid_size ** 2) / 2):
            grid_imgs.append(example)
        else:
            grid_imgs.append(system.view(example))
    
   
    fig = plt.figure(figsize=(grid_size, grid_size))
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                     # creates grid_size x grid_size grid of axes
                     nrows_ncols=(grid_size, grid_size),
                     axes_pad=0,  # pad between axes in inch.
                     share_all=True,
                     )

    
    for ax, im in zip(grid, enumerate(grid_imgs)):
        i, im = im
        #ax.set_xticks([])
        #ax.set_yticks([])
        ax.tick_params(bottom = False, left = False)
        # Remove batch dim and put channel dim last.
        ax.imshow(im[0].detach().permute(1, 2, 0))
    
    grid[0].get_yaxis().set_ticks([])
    grid[0].get_xaxis().set_ticks([])
    plt.axis('off')
    plt.savefig('./gridimgnew/audiomnist_viewmaker_' + str(img_idx) + '.jpg',bbox_inches='tight', pad_inches=0.0)
    plt.show()
    plt.clf()
    
    grid_imgs_diff = []
    for _ in range(grid_size ** 2):
        if _ == math.floor((grid_size ** 2) / 2):
            grid_imgs_diff.append(example)
        else:
            grid_imgs_diff.append(example - system.view(example))
    
    fig1 = plt.figure(figsize=(grid_size, grid_size))
    grid1 = ImageGrid(fig1, 111,  # similar to subplot(111)
                     # creates grid_size x grid_size grid of axes
                     nrows_ncols=(grid_size, grid_size),
                     axes_pad=0,  # pad between axes in inch.
                     share_all=True,
                    )

    cmap = plt.get_cmap('bwr_r')
    cmaplist = [cmap(i) for i in range(cmap.N)]
    cmap = cmap.from_list('Custom cmap', cmaplist, cmap.N)
    low = 0
    high = 0
    for i, im in enumerate(grid_imgs_diff):
        if i != math.floor((grid_size ** 2) / 2):
            low = min([low, torch.min(im).detach()])
            high = max([high, torch.max(im).detach()])
        
    bounds = np.arange(low, high, .5)
    idx=np.searchsorted(bounds,0)
    bounds=np.insert(bounds,idx,0)
    norm = BoundaryNorm(bounds, cmap.N)

    for ax, im in zip(grid1, enumerate(grid_imgs_diff)):
        i, im = im
        #ax.set_xticks([])
        #ax.set_yticks([])
        ax.tick_params(bottom = False, left = False)
        # Remove batch dim and put channel dim last.
        if i == math.floor((grid_size ** 2) / 2):
            ax.imshow(im[0].detach().permute(1, 2, 0))
        else:
            ax.imshow(im[0].detach().permute(1, 2, 0), norm=norm, cmap=cmap)
    
    grid1[0].get_yaxis().set_ticks([])
    grid1[0].get_xaxis().set_ticks([])
    plt.axis('off')
    plt.savefig('./gridimgnew/audiomnist_viewmakerdiff_' + str(img_idx) + '.jpg',bbox_inches='tight', pad_inches=0.0)
    plt.show()
    plt.clf()
    
    grid_imgs_expert = []
    for _ in range(grid_size ** 2):
        if _ == math.floor((grid_size ** 2) / 2):
            grid_imgs_expert.append(example)
        else:
            #grid_imgs_expert.append(example - dataset2[img_idx][1])
            grid_imgs_expert.append([dataset2[img_idx][1]])
    
   
    fig2 = plt.figure(figsize=(grid_size, grid_size))
    grid2 = ImageGrid(fig2, 111,  # similar to subplot(111)
                     # creates grid_size x grid_size grid of axes
                     nrows_ncols=(grid_size, grid_size),
                     axes_pad=0,  # pad between axes in inch.
                     share_all=True,
                     )

    
    for ax, im in zip(grid2, enumerate(grid_imgs_expert)):
        i, im = im
        #ax.set_xticks([])
        #ax.set_yticks([])
        ax.tick_params(bottom = False, left = False)
        # Remove batch dim and put channel dim last.
        ax.imshow(im[0].detach().permute(1, 2, 0))
    
    grid2[0].get_yaxis().set_ticks([])
    grid2[0].get_xaxis().set_ticks([])
    plt.axis('off')
    plt.savefig('./gridimgnew/audiomnist_expertspectral_' + str(img_idx) + '.jpg',bbox_inches='tight', pad_inches=0.0)
    plt.show()
    plt.clf()
```
