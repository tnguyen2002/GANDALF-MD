```python
train_cell_types = []
for cur_train_mask in train_masks:
    cur_train_mask = cur_train_mask.reshape(65536, 6)
    cell_type_counts = np.sum(cur_train_mask, axis=0)
    cell_type = np.argmax(cell_type_counts)
    train_cell_types.append(cell_type)
train_cell_types = np.array(train_cell_types)
```

```python
val_cell_types = []
for cur_train_mask in val_masks:
    cur_train_mask = cur_train_mask.reshape(65536, 6)
    cell_type_counts = np.sum(cur_train_mask, axis=0)
    cell_type = np.argmax(cell_type_counts)
    val_cell_types.append(cell_type)
val_cell_types = np.array(val_cell_types)
```

```python
test_cell_types = []
for cur_train_mask in test_masks:
    cur_train_mask = cur_train_mask.reshape(65536, 6)
    cell_type_counts = np.sum(cur_train_mask, axis=0)
    cell_type = np.argmax(cell_type_counts)
    test_cell_types.append(cell_type)
test_cell_types = np.array(test_cell_types)
```

```python
side_dict = {'left': 0, 'right': 1}
df = pd.read_csv("/data5/xiluohe/diabetic-retinopathy-detection/trainLabels.csv")
df["side_label"] = df["side"]
df.replace({'side_label': side_dict}, inplace=True)
from sklearn.utils import shuffle
df = shuffle(df)
df.reset_index(drop=True, inplace=True)

df['keep'] = df.index
df['keep'][df.index <= 1500] = 1
df['keep'][df.index > 1500] = 0

df = df[(df['keep']==1)]
df.reset_index(drop=True, inplace=True)

df['split'] = df.index
df['split'][df.index <= 1199] = "train"
df['split'][df.index > 1199] = "val"
display(df)
train_df = df[(df['split']=='train')]
val_df = df[(df['split']=='val')]
display(train_df)
train_df.to_csv("/data5/xiluohe/diabetic-retinopathy-detection/train_1500.csv", index=False)
val_df.to_csv("/data5/xiluohe/diabetic-retinopathy-detection/val_1500.csv", index=False)
```
